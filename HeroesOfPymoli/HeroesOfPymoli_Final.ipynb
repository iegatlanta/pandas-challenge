```python
#dependencies
import pandas as pd

#load csv file
purchase_data = "Resources/purchase_data.csv"

#read file and store in df
purchase_data_df = pd.read_csv(purchase_data)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
#calculate player count 
index = purchase_data_df.index
Total_Players = len(index)
Total_Players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#calculate number of unique items in df
item_name = len(purchase_data_df["Item Name"].unique())
purchase_data_df["Total Revenue"] = purchase_data_df.iloc[:, 7:].sum(axis=1)
total_revenue = purchase_data_df["Price"].sum()
price = purchase_data_df["Price"].mean()

total_summary_df = pd.DataFrame({"Number of Unique Items": [item_name],"Average Price": price,
"Number of Purchases": (Total_Players), "Total Revenue": total_revenue})

pd.options.display.float_format = '${:,.2f}'.format
total_summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#create variables
gender_totals = purchase_data_df["Gender"].value_counts()
player_count = len(index)
gender_percentage = gender_totals / player_count * 100
gender_breakdown = pd.DataFrame({"Total Count": gender_totals, 
                                 "Percentage of Players": gender_percentage})
#format
pd.options.display.float_format = '{:,.2f}%'.format
gender_breakdown
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#GroupBy will separate the data according to "Gender" values
gender_groupby_purchased_df = purchase_data_df.groupby(["Gender"])
gender_groupby_purchased_df["Purchase ID"].count().head()

#Average purchase price by gender
avg_purchase_price = gender_groupby_purchased_data_df["Price"].mean()
avg_purchase_price.head()
avg_purchase_price = avg_purchase_price.map("${:,.2f}".format)

#total purchase value by gender
total_value = gender_groupby_purchased_data_df["Price"].sum()
total_value.head()
new_total_value = total_value.map("${:,.2f}".format)

#total purchase value divided by purchase count by gender and organize summary gender data
norm_totals = total_value/gender_groupby_purchased_data_df["Purchase ID"].count()
norm_totals = norm_totals.map("${:,.2f}".format)
gender_purchase_df = pd.DataFrame(gender_groupby_purchased_data_df["Purchase ID"].count())
gender_purchase_df["Average Purchase Price"] = avg_purchase_price  
gender_purchase_df["Total Purchase Value"] = new_total_value 
gender_purchase_df["Avg Total Purchase per Person"] = norm_totals 

#summary analysis of purchasing DF grouped by gender, rename "Purchase ID" column, using .rename(columns={})
summary_gender_purchase_df = gender_purchase_df.rename(columns={"Purchase ID":"Sale Count"})
summary_gender_purchase_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99]

#create names for the bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

grp_by_age_purchase_data_df = purchase_data_df
grp_by_age_purchase_data_df["Age Group"] = pd.cut(grp_by_age_purchase_data_df["Age"], age_bins, labels=group_names)

#creating a group based the bins
grp_by_age_purchase_data_df = grp_by_age_purchase_data_df.groupby("Age Group")
grp_by_age_purchase_data_df.count()

#create new dataframe
summary_by_age_df = pd.DataFrame(grp_by_age_purchase_data_df.count())

#"purchase Id" column calculations and format percentages
summary_by_age_df["Purchase ID"] = (summary_by_age_df["Purchase ID"]/Total_Players)*100
summary_by_age_df["Purchase ID"] = summary_by_age_df["Purchase ID"].map("{:,.2f}%".format)

#reduce number of columns to "Age Summary", "Purcahe ID", format percentages and write into org Data Frame
org_summary_by_age_df = summary_by_age_df[["Purchase ID","SN"]]

#Rename columns
age_group_summary_df = org_summary_by_age_df.rename(columns={"SN":"Total Count", "Purchase ID":"Percentage of Players"})
age_group_summary_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#sort by age df
analysis_by_age_df = pd.DataFrame(grp_by_age_purchase_data_df["Purchase ID"].count())

#get Total purchase value by age
total_purchase_value_age = grp_by_age_purchase_data_df["Price"].sum()
total_purchase_value_age
dlr_total_purchase_value_age = total_purchase_value_age.map("${:,.2f}".format)

#get Average purchase price by age
avg_purchase_price_age = grp_by_age_purchase_data_df["Price"].mean()
avg_purchase_price_age
dlr_avg_purchase_price_age = avg_purchase_price_age.map("${:,.2f}".format)

#get Normalized totals by age
normalized_totals_age = total_purchase_value_age/grp_by_age_purchase_data_df["Purchase ID"].count()
dlr_normalized_totals_age = normalized_totals_age.map("${:,.2f}".format)

#organize summary gender data
analysis_by_age_df["Average Purchase Price"] = dlr_avg_purchase_price_age  
analysis_by_age_df["Total Purchase Value"] = dlr_total_purchase_value_age 
analysis_by_age_df["Avg Total Purchase per Person"] = dlr_normalized_totals_age 

#summary purchasing analysis df grouped by age, rename columns
summary_age_purchased_data_df = analysis_by_age_df.rename(columns={"Purchase ID":"Purchase Count"})
summary_age_purchased_data_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#original purchase data
orig_purchase_data_df = pd.DataFrame(purchase_data_df)

#group by Spendors ( "SN" )
grp_SN_top_cust_df = orig_purchase_data_df.groupby("SN")

#sort data by SN new dataframe and get Total purchase value
analysis_by_cust_df = pd.DataFrame(grp_SN_top_cust_df["Purchase ID"].count())
total_purchase_value_SN = grp_SN_top_cust_df["Price"].sum()

#get Average purchase price by SN and Reset summary Top Spender data
avg_purchase_price_SN = grp_SN_top_cust_df["Price"].mean()
avg_purchase_price_SN
dlr_avg_purchase_price_SN = avg_purchase_price_SN.map("${:,.2f}".format)
analysis_by_cust_df["Average Purchase Price"] = dlr_avg_purchase_price_SN
analysis_by_cust_df["Total Purchase Value"] = total_purchase_value_SN 

#summary Top Spendor analysis grouped by SN and rename "Purchase ID"
SUM_SN_purchased_data_df = analysis_by_cust_df.rename(columns={"Purchase ID":"Purchase Count"})
top_cust_df=SUM_SN_purchased_data_df.sort_values("Total Purchase Value", ascending=False)

#format Total Purchase Price
dlr_total_purchase_value_SN = total_purchase_value_SN.map("${:,.2f}".format)
top_cust_df["Total Purchase Value"] = dlr_total_purchase_value_SN
top_cust_df.head()
```
